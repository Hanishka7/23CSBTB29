# **Implement Basic Data Structures using Numpy, Pandas**

```python
import numpy as np

set1 = np.array([1, 2, 3, 4, 5])
set2 = np.array([10, 20, 30, 40, 50])

multiplication = set1 * set2
division = set1 / set2
power_off = set1 ** 2

print("Multiplication:", multiplication)
print("Division:", division)
print("Power Off (Squared):", power_off)
```

```python
import numpy as np
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])
result = array1 + array2
output = "Addition of two arrays: {} + {} = {}".format(array1, array2, result)
print(output)

```

```python
import numpy as np
array = np.array([1, 2, 3, 4, 5])
sin_result = np.sin(array)
log_result = np.log(array)
log2_result = np.log2(array)
exp_result = np.exp(array)
output_sin = "np.sin({}) = {}".format(array, sin_result)
output_log = "np.log({}) = {}".format(array, log_result)
output_log2 = "np.log2({}) = {}".format(array, log2_result)
output_exp = "np.exp({}) = {}".format(array, exp_result)

print(output_sin)
print(output_log)
print(output_log2)
print(output_exp)
```

## **visualization of Data using matplotlib, pyplots Packages[CO2]**

```python
import pandas as pd
data = {
    'Name': ['Kiran', 'Arun', 'Vijay', 'Varun'],
    'Age': [25, 30, 35, 40],
    'Height': [145, 151, 165, 173],
    'Weight': [45, 55, 65, 75]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Weight'], color='blue')
plt.title('Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.bar(df['Name'], df['Height'], color='green')
plt.title('Height of Individuals')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()
```

```python
plt.hist(df['Age'], bins=5, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
